```python
covid['Date'] = pd.to_datetime(covid['Date'], format = '%Y-%m-%d')
```

```python
#Active Cases
covid['Active_case']= covid['Confirmed']-(covid['Cured']+ covid['Deaths'])
covid.tail()
```

```python
statewise= pd.pivot_table(covid,values=["Confirmed","Deaths","Cured"],index = "State/UnionTerritory", aggfunc = max)
print(statewise)
```

```python
statewise["Recovery Rate"] = statewise["Cured"]*100/statewise["Confirmed"] #Recovery rate
```

```python
statewise["Mortality Rate"] = statewise["Deaths"]*100/statewise["Confirmed"] #Mortality rate
```

```python
statewise= statewise.sort_values(by = "Confirmed",ascending = False)
```
