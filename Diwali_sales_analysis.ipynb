```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


diwali= pd.read_csv(r"C:\Users\zendservices\Downloads\Python_Diwali_Sales_Analysis\Python_Diwali_Sales_Analysis\Diwali Sales Data.csv",encoding='iso-8859-1')
diwali.head()
```

```python
diwali.info()
```

```python
diwali.describe()
```

```python
diwali.isnull().sum()
```

```python
diwali.drop(["Status","unnamed1"],inplace=True, axis=1)
```

```python
diwali.isnull().sum()
```

```python
diwali.dropna(inplace=True)
```

```python
diwali['Amount']= diwali['Amount'].astype(int)
```

```python
diwali.rename(columns ={'Marital_Status':'Shadii'})
```

```python
diwali.columns
```

```python
#male kitne hai or female kitne hai
Counting = diwali.groupby(by = "Gender").size()[['M','F']].reset_index()
print(Counting.values)

ax= sns.countplot(x='Gender',data=diwali)
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
Gender_pay = diwali.groupby("Gender").sum()['Amount'].sort_values(ascending=False).reset_index()
print(Gender_pay)
ax= sns.barplot(x='Gender',y='Amount', data=Gender_pay)
plt.show()
```

```python
ax=sns.countplot(x='Age Group',data=diwali,hue= 'Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
amount_count_age_group=diwali.groupby(by='Age Group').sum()['Amount'].sort_values(ascending=False).reset_index()
ax=sns.barplot(x='Age Group',y='Amount',data=amount_count_age_group,linewidth=2,edgecolor="Black")
for bars in ax.containers:
    ax.bar_label(bars)
    plt.show()
```

```python
#total number of orders from top 10 states
top_10_states = diwali.groupby(by = 'State').sum()['Orders'].sort_values(ascending=False).reset_index()
plt.figure(figsize=(25,10))
sns.barplot(x='State',y='Orders',data=top_10_states)
plt.show()
```

```python
# Martial status
plt.figure(figsize=(10,5))
ax=sns.countplot(data=diwali,x='Marital_Status')

for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
Expense_data = diwali.groupby(by=['Marital_Status','Gender']).sum()['Amount'].sort_values(ascending=False).reset_index()
ax=sns.barplot(x='Marital_Status',y='Amount',hue='Gender',data=Expense_data)
for bars in ax.containers:
    ax.bar_label(bars, fmt='%d') # Set the format to integer (%d)
plt.show()
```

```python
plt.figure(figsize=(15,9))
ax=sns.countplot(x='Occupation',data= diwali,hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars ,fmt='%d')
plt.show()
```

```python
Expense_data=diwali.groupby(by='Occupation').sum()['Amount'].sort_values(ascending=False).reset_index()
plt.figure(figsize=(15,9))
ax=sns.barplot(x='Occupation',y='Amount',data = Expense_data,)
for bars in ax.containers:
    ax.bar_label(bars,fmt='%d')
    plt.show()
```

```python
#product
plt.figure(figsize=(25,5))
ax=sns.countplot(x='Product_Category',data=diwali)
for bars in ax.containers:
    ax.bar_label(bars,fmt='%d')
```

```python
plt.figure(figsize=(35,12))
product_cate= diwali.groupby(by='Product_Category').sum()['Amount'].sort_values(ascending=False).reset_index()
ax=sns.barplot(x='Product_Category',y='Amount',data=product_cate)
for bars in ax.containers:
    ax.bar_label(bars,fmt=('%d'))
```

```python
plt.figure(figsize=(20,5))
sales_data= diwali.groupby(by='Product_ID').sum()['Orders'].sort_values(ascending=False).reset_index().head(10)
ax=sns.barplot(x='Product_ID',y='Orders',data=sales_data)
for bars in ax.containers:
    ax.bar_label(bars,fmt=('%d'))
plt.show()
```
